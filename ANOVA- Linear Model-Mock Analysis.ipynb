```r
#The goal of this notebook is to generate random normally distributed data for two groups.
```

```r
#Two sample groups, ctrl and trt
#Trt is another bell curvve shifted by 5
library(ggplot2)

ctrl<-rnorm(5)
trt<-rnorm(5, mean=5)


df<-matrix(c(ctrl,trt), nrow=5,ncol=2) #Combine samples to one vector


labels<-c('ctrl','trt')
colnames(df)<-labels


df<-as.data.frame(df) 
df

stack_df<-stack(df) #Stack format to use for AOV 
stack_df
```

```r
#Visualize distribution with boxplot
plot<-ggplot(stack(df[1:2]), aes(x = ind, y = values)) + 
  geom_boxplot(notch=FALSE, fill='#aa22cc', color="black")+
    labs(title="Visualizing ctrl and trt distributions", x ="Group", y = "Value")
plot
```

```r
#ANOVA analysis
summary(aov(formula=values~ind, data=stack_df)) #1.09e-05 small p value indicate significant difference
```

```r
#Linear Model analysis
df_lm<-lm(formula=values~ind, data=stack_df)
summary(df_lm) #1.09e-05 p value indicates significant difference
```
