The purpose of this notebook is do perform statistical analysis on a trasncriptome dataset of mouse genes. 

```r
#Load dataframe from csv
df=read.csv('/Users/danielaquijano/Documents/GitHub/Transcriptomics-Course-Projects/mousegenes.csv')
head(df)
```

```r
#Find the length of each gene by subtracting Start from End range
#Add new Column to df
Length<-df$End-df$Start
df<-cbind(df, Length)
head(df)
```

```r
#ggplot library, visualize frequencies of gene lengths
#For better visualization, changing the number of bins to 80
library(ggplot2)
plot<-ggplot(df, aes(x=Length)) + 
  geom_histogram(bins=80, color="black", fill="pink")+
    geom_vline(aes(xintercept=mean(Length)),color=("purple"),linetype="dashed")+
        geom_vline(aes(xintercept=median(Length)),color='green',linetype="dashed")+theme_minimal()+
            labs(title="Gene Length Frequency", x ="Gene Length (bp)", y = "Frequency")

plot
```

```r
#plot log of the length in order to obtain normal-like distribution
#The vertical lines show that mean is close to median as is the case in an almost normal distribution
log_length<-log2(df$Length)
log2_mean<-mean(log_length)
log2_median<-median(log_length)
plot2<-ggplot(df, aes(x=log_length)) + 
  geom_histogram(bins=80, color="black", fill="pink")+
    geom_vline(aes(xintercept=log2_mean),color=("purple"),linetype="dashed")+
        geom_vline(aes(xintercept=log2_median),color='green',linetype="dashed")+theme_minimal()+
            labs(title="Log2 (Gene Length Frequency)", x ="Log2(Gene Length (bp))", y = "Frequency")
plot2
```

```r
#Add log_length column to df
#Finding the IQR of Log_length
df<-cbind(df,log_length)
head(df)
IQR(log_length)
quantile(log_length)
```

```r
#Exclude values outside the IQR and calculate mean and median once again 
#Exclude values greater than 75%, and less than 25%
iqr<-subset(df,log_length<=quantile(df$log_length)[4])
iqr<-subset(iqr,log_length>=quantile(df$log_length)[2])
mean(iqr$log_length)
median(iqr$log_length) #Mean and median are now almost equal
```

```r
#Calulate standard deviation of log_length
#Plot log_length
sd(df$log_length)
plot3<-ggplot(df, aes(x='log_length', y=log_length)) + 
  geom_boxplot(notch=FALSE, fill='#AC88FF', color="black")+
    labs(title="Log2 (Gene Length Frequency)", x ="Log2(Gene Length (bp))", y = "Log2(Gene Length (bp))")
plot3
```

```r
#Look at variation in gene lengths chromosome by chromosome
#Boxplot with separate boxplot for each chromosome
plot3<-ggplot(df, aes(x=Chr, y=log_length, fill=Chr)) + 
  geom_boxplot(color="black")+
    labs(title="Log2 (Gene Length) by chromosome", x ="Chromosome Number", y = "Log2 (Gene Length)")+
        scale_colour_manual(values = rainbow(21))
plot3
```

```r
#Randomly sample with no replacaement 100 genes and calculate mean log_length for the sample,repeat 1000 times
means_100<-numeric()

for (i in 1:100){
    
    randomselect=sample(nrow(df),100,replace=F)
    means_100[i]=mean(df[randomselect,'log_length'])
}
```

```r
#Count the frequency of each mean value obtained from each sample of 100 genes, compare to known population mean 
#Purple line is the mean, green is the median, they overlap, means they are very close together
plot4<-ggplot(as.data.frame(means_100), aes(means_100)) + 
  geom_histogram(bins=12, color="black", fill="pink")+
    geom_vline(aes(xintercept=mean(means_100)),color=("purple"),linetype="dashed")+
        geom_vline(aes(xintercept=median(means_100)),color='green',linetype="dashed")+theme_minimal()+
            labs(title="Frequency of mean values of 1000 samples of 100 values", x ="Mean n=100, Log(Gene Length)", y = "Frequency")
plot4
```

```r
#Calcualte standard deviation of sample means 
sd(means_100)
```

```r
#Repeat sampling process, but this time use n=1000 sample size

means_1000<-numeric()

for (i in 1:100){
    
    randomselect=sample(nrow(df),1000,replace=F)
    means_1000[i]=mean(df[randomselect,'log_length'])
}

plot4<-ggplot(as.data.frame(means_1000), aes(means_1000)) + 
  geom_histogram(bins=12, color="black", fill="pink")+
    geom_vline(aes(xintercept=mean(means_1000)),color=("purple"),linetype="dashed")+
        geom_vline(aes(xintercept=median(means_1000)),color='green',linetype="dashed")+theme_minimal()+
            labs(title="Frequency of mean values of 1000 samples of 100 values", x ="Mean n=100, Log(Gene Length)", y = "Frequency")
plot4
sd(means_1000) #Standard deviation decreases with increasing sample size
```
