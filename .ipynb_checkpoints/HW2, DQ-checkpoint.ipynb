```r
#Read in file as a table. Print first few rows
expvalues<-read.table('/Users/danielaquijano/Documents/NYU MASTERS/SPRING 2022/Transcriptomics/Source_Files/expvalues.txt',sep='')
#head(data)
expvalues<-as.matrix(expvalues)
head(expvalues)
```

```r
#Write function to calculate log of ratio of experimental group to control group
#These are the groups/factors
#The output is the dataframe with a new column appended to it that contains the mean of the factors, the ratio between them and the log of the ratio
expgroups<-factor(c('control','control','control','experimental','experimental','experimental'))

getlogFC<-function(expvalues, expgroups){
    avg<-vector()

    for (i in 1:nrow(expvalues)){
    x<-tapply(as.numeric(expvalues[i,]),expgroups,mean)
    avg<-rbind(avg,x)
    }

    rownames(avg)<-rownames(expvalues)
    expvalues<-cbind(expvalues, avg)
    ratio<-expvalues[,8]/expvalues[,7]
    expvalues<-cbind(expvalues, ratio)
    logcalc<-log2(expvalues[,9])
    expvalues<-cbind(expvalues, logcalc)
    return (expvalues)
   
}
```

```r
expvalues<-getlogFC(expvalues, expgroups) #Call the function, assign function output to dataframe
```

```r
#Setting var.equal to true since we are assuming that the variance is the same
#Function to calcualte T test p value. 
#Add column with p value to dataframe
getTtestPvalue<-function(expvalues, expgroups ){
    pval<-vector()

    for (i in 1:nrow(expvalues)){
    x<-t.test(expvalues[i,1:3],expvalues[i,4:6],var.equal=T)$p.value
    pval<-rbind(pval,x)
    }
    rownames(pval)<-rownames(expvalues)
    expvalues<-cbind(expvalues, pval)
    colnames(expvalues)[11]<-"P-values"
}



```

```r
head(expvalues)
nrow(expvalues) #The total number of genes is 22810
```

```r
#Count the number of genes that have a p-value<0.05
filter<-as.data.frame(subset(expvalues, expvalues[,11] < 0.05))
head(filter)
sprintf(' The number of genes with a p value less than 0.05 is %s', nrow(filter))
```

```r
#Filter out values based on log ratio greater than 1 or less than negative 1 (abs>1)
filter2<-as.data.frame(subset(expvalues, expvalues[,10] > 1 | expvalues[,10] < -1))
sprintf(' The number of genes with a log ratio with absolute value above 1 is %s', nrow(filter2))
```

```r
#Use logical operators to filter rows that have both a p value less than 0.05 and log change absolute value greater than 1
filter3<-as.data.frame(subset(expvalues, expvalues[,11] < 0.05 & expvalues[,10] > 1 | expvalues[,10] < -1))
sprintf(' The number of genes with a log ratio with absolute value above 1 and with a p value less than 0.05 is %s', nrow(filter3))
```
