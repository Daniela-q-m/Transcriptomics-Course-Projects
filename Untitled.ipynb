```r
library(ggplot2)
df=read.csv('/Users/danielaquijano/Documents/GitHub/Transcriptomics-Course-Projects/mousegenes.csv')
head(df)
plot<-ggplot(df, aes(x=Start)) + 
  geom_histogram(bins=80, color="black", fill="pink")+
    geom_vline(aes(xintercept=mean(Start)),color=("purple"),linetype="dashed")+
        geom_vline(aes(xintercept=median(Start)),color='green',linetype="dashed")+theme_minimal()+
            labs(title="Gene Length Frequency", x ="Gene Length (bp)", y = "Frequency")


plot
```
